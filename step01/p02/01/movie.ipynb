```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

from kmr_dataset import load_rates, get_paths
```

```python
# kmrd

path = get_paths(size='small')[3]
df_movies = pd.read_csv(path, sep='\t').dropna()
df_movies['year'] = df_movies['year'].astype('int')

path = get_paths(size='small')[2]
df_genres = pd.read_csv(path, sep=',').dropna()

path = get_paths(size='small')[0]
df_castings = pd.read_csv(path, sep=',').dropna()

path = get_paths(size='small')[4]
df_people = pd.read_csv(path, sep='\t').dropna()

path = get_paths(size='small')[1]
df_countries = pd.read_csv(path, sep=',').dropna()

path = get_paths(size='small')[5]
df_rates = pd.read_csv(path, sep=',').dropna()

df_castings = pd.merge(df_castings, df_people, on=['people'], how='inner')
df_castings.drop(columns=['people', 'order', 'leading'], inplace=True)

merged_df = pd.merge(df_movies, df_genres, on=['movie'], how='inner')
merged_df = merged_df.groupby(['movie', 'title', 'title_eng', 'year', 'grade']).agg({'genre': ', '.join}).reset_index()

merged_df = pd.merge(merged_df, df_castings, on=['movie'], how='inner')
merged_df = merged_df.groupby(['movie', 'title', 'title_eng', 'year', 'grade', 'genre']).agg({'korean': ', '.join}).reset_index()

merged_df = pd.merge(merged_df, df_countries, on=['movie'], how='inner')
merged_df = merged_df.groupby(['movie', 'title', 'title_eng', 'year', 'grade', 'genre', 'korean']).agg({'country': ', '.join}).reset_index()

merged_df.head(10)
```

```python
tmp = pd.merge(df_movies, df_rates, on=['movie'], how='inner')
tmp = tmp.groupby(['title', 'title_eng', 'year']).agg({'rate': np.mean}).reset_index()
tmp.head(10)
```

```python
# Netflix

import csv

input_file = './Netflix/movie_titles.txt'
output_file = './Netflix/movie_titles_preprocessed.txt'

df_movies = []

with open(input_file, 'r', encoding='ISO-8859-1') as infile:
    reader = csv.reader(infile)

    for row in reader:
        if len(row) > 3:
            row[2] = ','.join(row[2:])
        df_movies.append(row[:3])

df_movies = pd.DataFrame(df_movies, columns=['movie', 'year', 'title'])
df_movies.head(10)
```

```python
df_rate_mv_0000001 = pd.read_csv('./Netflix/rates/mv_0000001.txt', sep=',', skiprows=1, names=['user', 'rate', 'date'])
df_rate_mv_0000001.head(10)
```

```python
df_rates = []

for i in range(1, 11):
    df_rate = pd.read_csv(f'./Netflix/rates/mv_{i:07}.txt', sep=',', skiprows=1, names=['user', 'rate', 'date'])
    mean = df_rate['rate'].mean()
    df_rates.append([f'{i}', mean])
    
df_rates = pd.DataFrame(df_rates, columns=['movie', 'rate'])
df_movies = df_movies[:10]

df_merged = pd.merge(df_movies, df_rates, on=['movie'], how='inner')
df_merged.head(10)
```

```python
#MovieLens

df_movies = pd.read_csv('./MovieLens/movies.csv').dropna()
df_movies.head(10)
```

```python
df_tags = pd.read_csv('./MovieLens/tags.csv').dropna()
df_tags = df_tags.groupby(['movieId']).agg({'tag': lambda x: ', '.join(set(x))}).reset_index()

df_merged = pd.merge(df_movies, df_tags, on=['movieId'], how='inner')
df_merged.head(10)
```

```python
df_ratings = pd.read_csv('./MovieLens/ratings.csv').dropna()
df_ratings = df_ratings.groupby(['userId']).agg({
    'rating': np.mean, 
    'movieId': lambda x: ', '.join(sorted(map(str, set(x))))
}).reset_index()
df_ratings.head(10)
```
