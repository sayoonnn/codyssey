- 영화의 장르별 분포 (Bar chart) 및 장르별 평균 평점을 확인하라.

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from matplotlib.ticker import MultipleLocator
import pandas as pd

plt.rcParams['font.family'] = 'D2coding'
plt.rcParams['font.size'] = 9
plt.rcParams['axes.unicode_minus'] = False
```

```python
KMRD_DATASET_PATH = '../../../kmrd-small/'

df_genre = pd.read_csv(KMRD_DATASET_PATH + 'genres.csv').dropna()
```

```python
tmp = df_genre.groupby('genre').size().sort_values(ascending=True)

print(tmp.count())

tmp.plot(kind='barh', width=.8)
```

```python
df_rates = pd.read_csv(KMRD_DATASET_PATH + 'rates.csv').dropna()

tmp2 = df_rates.merge(df_genre, on='movie').groupby('genre').agg({'rate': 'mean'}).sort_values(by='rate', ascending=True)


print(tmp2)


ax = tmp2.plot(kind='barh', width=.7)

```
