```python
import pandas as pd
import re
```

```python
df = pd.read_csv("data/prob-0101.csv")
```

```python
movies = df['제목'].drop_duplicates().values
publishers = set([publishers for p in df['배급사'] for publishers in re.split('[,]+', p)])
directors = df['감독'].drop_duplicates().values
actors = set([actor for t in df['출연진'].values for actor in t.split(', ')])
genre = set([genre for g in df['장르'] for genre in re.split('[/,·]+', g)])

print(f"영화: {len(movies)}")
print(f"배급사: {len(publishers)}")
print(f"감독: {len(directors)}")
print(f"출연진: {len(actors)}")
print(f"장르: {len(genre)}")
```

```python
# 장르 기준 추천
pub_recommend = df[df['장르'].str.contains('로맨스', regex=False, na=False)]
print("로맨스 영화")
print(pub_recommend)
print()

pub_recommend = df[df['장르'].str.contains('드라마', regex=False, na=False)]
print("SF 영화")
print(pub_recommend)
print()


pub_recommend = df[df['장르'].str.contains('액션', regex=False, na=False)]
print("액션 영화")
print(pub_recommend)
print()
```

```python
# 배급사 기준 추천
pub_recommend = df[df['배급사'].str.contains('넷플릭스', regex=False, na=False)]
print("넷플릭스 제작")
print(pub_recommend)
print()

pub_recommend = df[df['배급사'].str.contains('CJ', regex=False, na=False)]
print("CJ 투자")
print(pub_recommend)
print()
```
